```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')


# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Combine the data into a single dataset
df1 = pd.merge(pd.read_csv(mouse_drug_data_to_load),pd.read_csv(clinical_trial_data_to_load) , how = 'inner', on = 'Mouse ID' )

# Drop the Drugs not in the study
#df1.shape
sd =['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
df1 = df1[df1['Drug'].isin(sd)]
#df1.shape

# Display the data table for preview

df1.head(15)


```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mtv = df1.groupby(['Drug','Timepoint']).mean().reset_index()
mtv.drop('Metastatic Sites', axis = 1, inplace = True)
mtv.rename(columns={'Tumor Volume (mm3)':'MeanTumorVol(mm3)'}, inplace = True)
mtv.head(10)

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stv = df1.groupby(['Drug','Timepoint']).sem().reset_index()
stv.drop(['Metastatic Sites','Mouse ID'], axis = 1, inplace = True)
stv.rename(columns={'Tumor Volume (mm3)':'SEMTumorVol(mm3)'}, inplace = True)
# Convert to DataFrame

# Preview DataFrame
stv.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mtv_rf= mtv.pivot('Timepoint', 'Drug', 'MeanTumorVol(mm3)')
stv_rf =stv.pivot('Timepoint', 'Drug', 'SEMTumorVol(mm3)')

# Preview that Reformatting worked
stv_rf
```

```python
#plot error bars
ax = mtv_rf.plot(figsize=(12,8), yerr = stv_rf, legend = False)
#reset color cycle so that the marker colors match
ax.set_prop_cycle(None)
#plot the markers
mtv_rf.plot(figsize=(14,8), style='^-', markersize=7, ax = ax)

ax.set_xlabel('Treatment Duration (Days)')
ax.set_ylabel("MeanTumorVol(mm3)")
ax.set_xlim(-0.5, 50)
plt.title('Mean Tumor Volume Over Time')
plt.legend(loc="best", fontsize="small", fancybox=True, shadow = True)
plt.grid(axis = 'y')
plt.tight_layout()

# Save the Figure
plt.savefig("../Images/mean_tumor_volume.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

# Convert to DataFrame
mms = df1.groupby(['Drug','Timepoint']).mean().reset_index()
mms.drop('Tumor Volume (mm3)', axis = 1, inplace = True)
mms.rename(columns={'Metastatic Sites':'MeanMetastaticSites'}, inplace = True)

# Preview DataFrame
mms.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

SemMs = df1.groupby(['Drug','Timepoint']).sem().reset_index()
SemMs.drop(['Tumor Volume (mm3)','Mouse ID'], axis = 1, inplace = True)
SemMs.rename(columns={'Metastatic Sites':'SEMMetastaticSites'}, inplace = True)

# Preview DataFrame
SemMs.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mms_rf= mms.pivot('Timepoint', 'Drug', 'MeanMetastaticSites')
SemMs_rf =SemMs.pivot('Timepoint', 'Drug', 'SEMMetastaticSites')

# Preview that Reformatting worked
mms_rf.head()
#SemMs_rf.head()
```

```python
# Generate the Plot (with Error Bars)
#plot error bars
ax = mms_rf.plot(figsize=(12,8), yerr = SemMs_rf, legend = False)
#reset color cycle so that the marker colors match
ax.set_prop_cycle(None)
#plot the markers
mms_rf.plot(figsize=(14,8), style='^-', markersize=7, ax = ax)

ax.set_xlabel('Treatment Duration (Days)')
ax.set_ylabel('Mean Metastatic Sites')
ax.set_xlim(-0.5, 50)
plt.legend(loc="best", fontsize="small", fancybox=True, shadow = True)
plt.title('Mean Metastatic Sites Over Time')
plt.grid(axis = 'y')
plt.tight_layout()

# Save the Figure
plt.savefig("../Images/mean_metastatic_sites.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

mcnt = df1.groupby(['Drug','Timepoint']).count().reset_index()
mcnt.drop(['Tumor Volume (mm3)','Metastatic Sites'], axis = 1, inplace = True)
mcnt.rename(columns={'Mouse ID':'MouseCount'}, inplace = True)

#get the starting mice count by using timepoint zero
mtot = mcnt.loc[mcnt['Timepoint'] == 0]
mtot.drop('Timepoint', axis = 1, inplace = True)
mtot.rename(columns={'MouseCount':'TotalMouse'}, inplace = True)

#merge the dataframes with mice count by timeperiod with starting mice count
srdf = pd.merge(mcnt, mtot, how = 'inner', on = 'Drug')

#create a survival rate for each drug and timepoint
srdf['Survival%'] = srdf['MouseCount']/srdf['TotalMouse']

# Preview DataFrame
#mcnt.head(10)
srdf.head(20)              
```

```python
# Minor Data Munging to Re-Format the Data Frames
srdf_rf= srdf.pivot('Timepoint', 'Drug', 'Survival%')
# Preview the Data Frame
srdf_rf.head(10)
```

```python
# Generate the Plot (Accounting for percentages)
srdf_rf.plot(figsize=(14,8), style='^-', markersize=7)

plt.xlabel('Treatment Duration (Days)')
plt.ylabel('% of Test Mice')
plt.title('Survival Rate')
ax.set_xlim(-0.5, 50)
plt.legend(loc="best", fontsize="small", fancybox=True, shadow = True)
plt.grid(axis = 'y')
plt.tight_layout()


# Save the Figure
plt.savefig("../Images/survival_rates.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#create series for timepoint 0 and timepoint 45
tvA = mtv.loc[mtv['Timepoint'] == 0]
tvA.drop('Timepoint', axis = 1, inplace = True)
tvB = mtv.loc[mtv['Timepoint'] == 45]
tvB.drop('Timepoint', axis = 1, inplace = True)

#merge the two series into dataframe and create % change column
tvPc = pd.merge(tvA, tvB, how = 'inner', on = 'Drug')
tvPc['TumorVolChange%'] = (tvPc['MeanTumorVol(mm3)_y'] - tvPc['MeanTumorVol(mm3)_x'])/tvPc['MeanTumorVol(mm3)_x'] 
tvPc.drop(['MeanTumorVol(mm3)_y', 'MeanTumorVol(mm3)_x'], axis = 1, inplace = True)
tvPc.sort_values(by=['TumorVolChange%'], inplace=True)
tvPc.reset_index(inplace = True, drop = True)

#create color list based decrease increase, list will be used to format bars
pf = ['green' if c < 0 else 'red' for c in tvPc['TumorVolChange%']]



tvPc.plot.bar(x='Drug', y = 'TumorVolChange%', color = pf ,figsize=(12,8), legend = False)
plt.grid(axis = 'y')
plt.ylabel('% Change in Tumor Vol')
plt.title('Tumor Response')

# Save the Figure
plt.savefig("../Images/tumor_response.png")
# Show the Figure
plt.show()
```
